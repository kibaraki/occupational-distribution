```python
path = '/content/drive/MyDrive/bias in LMs/constrained decoding/generations/'

names = ['gpt2', 'bert', 'ai-f', 'rinnNeo', 'rinn2', 'cl-t', 'Eleu']
folders = ['gpt2', 'mBERT', 'mGPT', 'neox-J', 'rinna-gpt2', 'tohoku-bert', 'neox']
prompts = ['job_is', 'job_type', 'job?']

for i in range(len(names)):
  tmp_path = path + folders[i] + '/' + names[i] + '_'
  tmp = pd.read_csv(tmp_path + prompts[0] + '_men.csv')
  tmp = tmp.drop(tmp.iloc[:, 4:], axis=1)

  for j in range(len(prompts)):
    full_path = tmp_path + prompts[j] + '_'
    men = pd.read_csv(full_path + 'men.csv')
    wom = pd.read_csv(full_path + 'wom.csv')


    men_tmp = men.drop(men.columns[[0, 1, 2, 3]],axis = 1)
    wom_tmp = wom.drop(wom.columns[[0, 1, 2, 3]],axis = 1)

    tmp['mean-men-' + names[i] + '_' + prompts[j]] = men_tmp.mean(axis=1)
    tmp['max-men-' + names[i] + '_' + prompts[j]] = men_tmp.max(axis=1)
    tmp['mean-wom-' + names[i] + '_' + prompts[j]] = wom_tmp.mean(axis=1)
    tmp['max-wom-' + names[i] + '_' + prompts[j]] = wom_tmp.max(axis=1)

  tmp.to_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/' + names[i] + '-mean-max.csv', index=False)
  tmp
```

```python
names = ['gpt2', 'bert', 'ai-f', 'rinnNeo', 'rinn2', 'cl-t', 'Eleu']
prompts = ['job_is', 'job_type', 'job?']

for i in range(len(names)):
  df = pd.read_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/extracted mean-max/' + names[i] + '-mean-max.csv')
  tmp = df.drop(df.iloc[:, 4:], axis=1)
  df['男'] = df['男'].replace('-', 0)
  df['女'] = df['女'].replace('-', 0)
  tmp['men_pt'] = df['男'].astype(float) / df['総数'].astype(float)
  tmp['wom_pt'] = df['女'].astype(float) / df['総数'].astype(float)
  for j in range(len(prompts)):
    tmp['mean_men_' + prompts[j] + '_pt'] = np.exp(df['mean-men-' + names[i] + '_' + prompts[j]]) / (np.exp(df['mean-men-' + names[i] + '_' + prompts[j]]) + np.exp(df['mean-wom-' + names[i] + '_' + prompts[j]]))
    tmp['max_men_' + prompts[j] + '_pt'] = np.exp(df['max-men-' + names[i] + '_' + prompts[j]]) / (np.exp(df['max-men-' + names[i] + '_' + prompts[j]]) + np.exp(df['max-wom-' + names[i] + '_' + prompts[j]]))

    tmp['mean_wom_' + prompts[j] + '_pt'] = np.exp(df['mean-wom-' + names[i] + '_' + prompts[j]]) / (np.exp(df['mean-men-' + names[i] + '_' + prompts[j]]) + np.exp(df['mean-wom-' + names[i] + '_' + prompts[j]]))
    tmp['max_wom_' + prompts[j] + '_pt'] = np.exp(df['max-wom-' + names[i] + '_' + prompts[j]]) / (np.exp(df['max-men-' + names[i] + '_' + prompts[j]]) + np.exp(df['max-wom-' + names[i] + '_' + prompts[j]]))

  tmp.to_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/' + names[i] + '_to_percent.csv', index=False)
```

```python
names = ['gpt2', 'bert', 'ai-f', 'rinnNeo', 'rinn2', 'cl-t', 'Eleu']
for i in range(len(names)):
	df = pd.read_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/' + names[i] + '_to_percent.csv')
	tmp = df.drop(df.iloc[:, 4:], axis=1)
	df['男'] = df['男'].replace('-', 0)
	df['女'] = df['女'].replace('-', 0)
	tmp['gold'] = df['女'].astype(float) - df['男'].astype(float)
	tmp['mean_job_is'] = (df['mean_wom_job_is_pt'] - df['mean_men_job_is_pt'])
	tmp['max_job_is'] = (df['max_wom_job_is_pt'] - df['max_men_job_is_pt'])
	tmp['mean_job_type'] = (df['mean_wom_job_type_pt'] - df['mean_men_job_type_pt'])
	tmp['max_job_type'] = (df['max_wom_job_type_pt'] - df['max_men_job_type_pt'])
	tmp['mean_job?'] = (df['mean_wom_job?_pt'] - df['mean_men_job?_pt'])
	tmp['max_job?'] = (df['max_wom_job?_pt'] - df['max_men_job?_pt'])

	tmp['gold_lab'] = np.where(tmp['gold'] > 0, 'f', 'm')
	tmp['mean_is_lab'] = np.where(tmp['mean_job_is'] > 0, 'f', 'm')
	tmp['max_is_lab'] = np.where(tmp['max_job_is'] > 0, 'f', 'm')
	tmp['mean_type_lab'] = np.where(tmp['mean_job_type'] > 0, 'f', 'm')
	tmp['max_type_lab'] = np.where(tmp['max_job_type'] > 0, 'f', 'm')
	tmp['mean?_lab'] = np.where(tmp['mean_job?'] > 0, 'f', 'm')
	tmp['max?_lab'] = np.where(tmp['max_job?'] > 0, 'f', 'm')

	label_only = tmp.drop(tmp.columns[[4, 5, 6, 7, 8, 9, 10]], axis=1)
	tmp.to_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/labels/' + names[i] + '_lab.csv', index=False)
	label_only.to_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/labels/' + names[i] + '_labonly.csv', index=False)
```

```python
# for all occ

names = ['gpt2', 'bert', 'ai-f', 'rinnNeo', 'rinn2', 'cl-t', 'Eleu']
lab = ['mean_is_lab',	'max_is_lab',	'mean_type_lab',	'max_type_lab',	'mean?_lab',	'max?_lab']

new_df = pd.DataFrame()


for i in range(len(names)):
  df = pd.read_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/labels/' + names[i] + '_labonly.csv')
  actual = df['gold_lab'].to_numpy()

  unique, counts = np.unique(actual, return_counts=True)
  print(dict(zip(unique, counts)))

  tmp_a = []
  for j in range(len(lab)):
    pred = df[lab[j]].to_numpy()
    mac = round(f1_score(actual, pred, average='macro'), 3)
    tmp_a.append(mac)
  new_df[names[i]] = np.array(tmp_a)

new_df.to_csv('/content/drive/MyDrive/bias in LMs/constrained decoding/all_occ.csv', index = False)

new_df
```
